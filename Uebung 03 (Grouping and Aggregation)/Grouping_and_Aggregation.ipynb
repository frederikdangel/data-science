### Demonstration of Aggregation: Simple case

```python
# select relevant subset of columns
h = op[['order_id','add_to_cart_order']]
h.head(15)
```

```python
aggregated_m = h.groupby(['order_id']).max() # Aggregation using MAX operator
aggregated_m.head(10)
```

```python
aggregated_s = h.groupby(['order_id']).size() # Aggregation using SIZE operator (=number of lines in each group)
aggregated_s.head(10)
```

### Demonstration of Aggregation: General case

#### Multiple columns and multiple aggregation operators per column

```python
# select relevant subset of columns
h = op[['order_id','add_to_cart_order','reordered']]
h.head()
```

```python
# group by order_id 
# using "size"=COUNT and "max" aggregation on add_to_cart_order
# and "mean" on reordered
aggregated = h.groupby(
    ['order_id']).agg({'add_to_cart_order' : [np.size, np.max],
                       'reordered' : np.mean
                     })
aggregated.head()
```

```python
print(type(aggregated))
print(aggregated.columns)
```
